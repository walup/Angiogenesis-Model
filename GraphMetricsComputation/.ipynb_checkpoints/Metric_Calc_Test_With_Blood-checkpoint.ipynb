## Blood Vessel Graph Test

```python
import numpy as np
from NetworkMetricCalculator import NetworkMetricCalculator
from NetworkMetricCalculator import NetworkMetricType
import matplotlib.pyplot as plt
import networkx as nx
```

```python
networkMetric = NetworkMetricCalculator()
adjacencyMatrix = networkMetric.loadAdjacencyMatrix('red_vacular_ejemplo.csv')
```

### Degree centrality

```python
#networkMetric.drawGraphFruchterman(adjacencyMatrix, 200, 1000, 1000, False)
degrees = networkMetric.computeNetworkMetric(adjacencyMatrix, NetworkMetricType.DEGREES)
averageDegree = networkMetric.computeNetworkMetric(adjacencyMatrix, NetworkMetricType.AVERAGE_DEGREE)

print("Average degree "+str(averageDegree))
values, degreeDist = networkMetric.computeDegreeDistribution(adjacencyMatrix)
plt.figure()
plt.title("Degree centralities")
plt.bar(list(range(0,len(degrees))),degrees, facecolor = "#f73181", edgecolor = "none")
plt.xlabel("Node")
plt.ylabel("Degree")

```

### Eccentricity

```python
eccentricity = networkMetric.computeNetworkMetric(adjacencyMatrix, NetworkMetricType.ECCENTRICITY)

print("Average eccentricity "+str(np.mean(eccentricity)))

plt.figure()
plt.title("Eccentricity")
plt.bar(list(range(0,len(eccentricity))),eccentricity, facecolor = "#f73181", edgecolor = "none")
plt.xlabel("Node")
plt.ylabel("Eccentricity")
```

### Clustering coefficients

```python
clusteringCoeffs = networkMetric.computeNetworkMetric(adjacencyMatrix, NetworkMetricType.CLUSTERING_COEFFICIENTS)
averageClusteringCoeff = networkMetric.computeNetworkMetric(adjacencyMatrix, NetworkMetricType.AVERAGE_CLUSTERING_COEFFICIENTS)

print("Average clustering coefficient "+str(averageClusteringCoeff))

plt.figure()
plt.title("Clustering Coefficient")
plt.bar(list(range(0,len(clusteringCoeffs))),clusteringCoeffs, facecolor = "#f73181", edgecolor = "none")
plt.xlabel("Node")
plt.ylabel("Clustering Coefficient")
```

### Page rank

```python
pageRank = networkMetric.computeNetworkMetric(adjacencyMatrix, NetworkMetricType.PAGE_RANK)
averagePageRank = networkMetric.computeNetworkMetric(adjacencyMatrix, NetworkMetricType.AVERAGE_PAGE_RANK)

print("Average page rank "+str(averagePageRank))

plt.figure()
plt.title("Page rank")
plt.bar(list(range(0,len(pageRank))),pageRank, facecolor = "#f73181", edgecolor = "none")
plt.xlabel("Node")
plt.ylabel("Page rank")
```
